```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('430889rows_30052024_with_outliers.csv', usecols=['tconst', 'titleType', 'genres',
                                                                  'primaryTitle', 'numVotes', 'averageRating', 
                                                                   'isAdult', 'startYear', 'runtimeMinutes'])
df.primaryTitle.duplicated().sum()
```

```python
# sort by numVote and startYear
df = df.sort_values(['numVotes', 'startYear'], ascending=False)

# Drop duplicates
df = df.drop_duplicates('primaryTitle')

df.primaryTitle.duplicated().sum()
```

```python
df.head(3)
```

```python
# Convert genres to list of strings
df['genres'] = df['genres'].str.split(',')

# Create numeric label for each category of genres
genres_cat = ['Action', 'Adventure', 'Animation', 'Biography','Comedy', 
               'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 
               'Horror','Mystery', 'Romance', 'Sci-Fi', 'Thriller',]

for i in genres_cat:
    df[i] = df['genres'].apply(lambda x:1 if i in x else 0)
```

```python
# Create numeric label for each category of titleType
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['type'] = label_encoder.fit_transform(df['titleType'])

# Drop non-selected columns
df = df.drop(columns=['genres', 'titleType', 'tconst'], axis = 1)
# Lower the title's names
df['primaryTitle']= df['primaryTitle'].apply(lambda x: x.lower())
```

```python
# Create a df to return values
df_return = df.copy()
df_return["Title"] = df_return["primaryTitle"]
df_return = df_return.set_index('primaryTitle')
df_return = df_return.rename({"averageRating": "Average Ratings", 
                              "numVotes": "Votes", "titleType": "Type of title",
                             "startYear": "Released year", "runtimeMinutes": "Duration",
                             "genres": "Genres", "isAdult": "Adult title"}, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# Initialize variable X
X = df.select_dtypes("number")

# Standardize features of X
scaler = StandardScaler()
scaler.fit(X)

X_scaled = pd.DataFrame(scaler.transform(X), index=X.index, columns=X.columns)
```

```python
#Create a function to return recommendated films

def get_recommendation(title):
    
    # Lower title's name
    title = title.lower()
    
    # To identify the searching film
    # Scale searching value
    input_title_scaled = X_scaled.loc[title].to_frame().T
    
    # Show the KPI of serching film to compare with returned values
    input_title = X.loc[title].to_frame().T
    


    # Fit model
    modelKNN = NearestNeighbors(n_neighbors=5)
    modelKNN.fit(X_scaled)

    # Identify distance and indeices of similar films
    neigh_dist, neigh_film = modelKNN.kneighbors(input_title_scaled, n_neighbors=11)
    film_similar = neigh_film[0][1:]
    print(df.loc[title]) 
    return df_return.iloc[film_similar]
```

```python
# Get the user input
title = input("Enter a movie title: ")

try:
    return get_recommendation(title)

except (KeyError, IndexError):
    print("sorry, that title does not exist. Please try another.")
    
```
